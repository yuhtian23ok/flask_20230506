```python
import requests

url = "https://raw.githubusercontent.com/roberthsu2003/PythonForDataAnalysis/master/%E6%AA%94%E6%A1%88%E5%AD%98%E5%8F%96/%E5%80%8B%E8%82%A1%E6%97%A5%E6%88%90%E4%BA%A4%E8%B3%87%E8%A8%8A.csv"
response = requests.get(url)

if response.ok:
    print("下載成功")
    print("下載內容如下:")
    content = response.text
else:
    print("下載失敗")
```

# 處理CSV

```python
#存檔
file = open('個股日成交資訊.csv',mode='w',encoding='utf-8',newline='')
file.write(content)
file.close()
```

```python
#處理csv檔
import csv
with open('個股日成交資訊.csv',encoding='utf-8',newline='') as file:
    csv_reader = csv.DictReader(file)
    #print(type(csv_reader))
    csv_list = list(csv_reader)
    
for item_dict in csv_list:
    #if item_dict['證券名稱'] == "富邦科技":
    #    print(item_dict)
    
    if "富邦" in item_dict['證券名稱']:
        print(item_dict)
```

```python
import requests
url = 'https://data.ntpc.gov.tw/api/datasets/71cd1490-a2df-4198-bef1-318479775e8a/csv?size=100'

response = requests.get(url)
if response.ok:
    print(response.text)
```

## 用 StringIO ，不用CSV存下資料

```python
from io import StringIO
import csv
file = StringIO(response.text,newline='')
csv_reader = csv.DictReader(file)
csv_list = list(csv_reader)
for item_dict in csv_list:
    print(item_dict)
```

# pandas: DataFrame

```python
import pandas as pd
df = pd.DataFrame(csv_list)
df
```

```python
df.drop(columns=['sareaen','snaen','aren','lat','lng'])
```
